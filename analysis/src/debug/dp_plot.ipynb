```python
def iMon_max(run_num):
    filename = f"../csv/run{run_num}_HViMon.csv"
    iMon = pd.read_csv(filename)
    
    print(len(iMon['comment_']))
    
    
    temp = iMon['comment_']
    descr = temp.drop_duplicates()
    print(len(descr))

    #print(descr)
    
    max_l = [0]
    names = ["MDT PS BC BOS2C08 ML2 HV Imon"]
    hasName_flag = False
    for j in iMon.index:
        com = iMon['comment_'][j]
        val = iMon['value_number'][j]
        for k in range(len(names)):
            if(com == names[k]):
                hasName_flag = True
                break
            else:
                hasName_flag = False
                
            
        if(not hasName_flag):
            names.append(com)
            max_l.append(val)
        else:
            ind = names.index(com)
            if(val > max_l[ind]):
                max_l[ind] = val
    print(len(max_l))
    return max_l
                    
                      
                    
                    
            
    
   
```

```python
#LUMINOSITY vs iMON
def iMon_lumiblock(run_num,descriptor):
    filename = f"../csv/run{run_num}_fsm.csv"
    fsm = pd.read_csv(filename)

    fsm_n = fsm[fsm['element_name'] == 'ATLMDTSCS:MDT_BA_RunInfo.RunParams.luminosityBlock'][['value_number','ts']]
    abbr = []
    for k in fsm.index:
        abbr.append('lumi')
    fsm_n['Abbr'] = pd.Series(abbr) 
    
    filename = f"../csv/run{run_num}_HViMon.csv"
    iMon = pd.read_csv(filename)
    iMon_n = iMon[iMon['comment_'] == f'{descriptor}'][['value_number','ts']]
    abbr = []
    for k in iMon.index:
        abbr.append('iMon')
    iMon_n['Abbr'] = pd.Series(abbr)
    
    db = pd.concat([fsm_n,iMon_n])
    db0 = db.sort_values(by = ['ts'], ascending = True)
    db0 = db0.reset_index()
    #print(db0[1270:])
    
    lumi_start_ts = 0
    iMon_prev_flag = 0
    iMon_temp = []
    iMon_val = []
    lumi = 0
    val_prev = 0
    for k in db0.index:
        Abbr = db0['Abbr'][k]
        ts = db0['ts'][k]
        val = db0['value_number'][k]
        if(Abbr == 'lumi' and val > (val_prev-15) and val < (val_prev + 15)):
            val_prev = val
            lumi_start_ts = ts
            iMon_val.append(iMon_temp[:])
        
            iMon_prev_flag = 0
            lumi = len(iMon_val)
            if(lumi > val):
                del iMon_val[-1]
            elif(lumi < val):
                dif = int(val-lumi)
                for k in range(dif):
                    iMon_val.append([])
            iMon_temp.clear()
        elif(Abbr == 'iMon'):
            iMon_temp.append(val)
            iMon_prev_flag = 1
    
    iMon_val.append(iMon_temp[:])
    print(lumi)
            
    return iMon_val
```

```python
def plot_iMon_lumi(run_num,iMon_val,name):
    filename = f"../csv/run{run_num}_lumi.csv"
    lumi = pd.read_csv(filename)
    #print(lumi)
    print()
    iMon_g =[]
    lumi_g = []
    lumi_l = []
    lumiBlock = []
    lumi_temp = []
    for j in lumi.index:
        lumi_l.append(lumi['LBAvInstLumi'][j])
        lumi_temp.append(lumi['Lumi_number'][j])
        
    #print(len(lumi_l))
    #print(len(iMon_val))
    x = array('d')
    y = array('d')
    for k in range(1,len(iMon_val)):
        if(len(iMon_val[k]) > 0):
            for j in range(len(iMon_val[k])):
                if(lumi_l[k] > 1.0):
                    iMon_g.append(iMon_val[k][j])
                    lumi_g.append(lumi_l[k])
                    lumiBlock.append(lumi_temp[k])
                
                
    #print(lumi_g)
    #print(len(lumi_g))
    #print(len(iMon_g))
    ##fig = plt.figure(figsize = (10,8))
    ##plt.scatter(lumi_g,iMon_g)
    ##plt.xlabel('Instantaneous Luminosity', fontsize='xx-large')
    ##plt.ylabel('iMon', fontsize='xx-large')
    #plt.title("Max iMon", fontsize='xx-large')
    c1 = rt.TCanvas( 'c1', 'Lumi_iMon', 200, 10, 700, 500 )
    c1.SetGrid()

    x, y = array( 'd' ), array( 'd' )
    for k in range(len(lumi_g)):
        x.append( lumi_g[k] )
        y.append( iMon_g[k])
    gr = rt.TGraph( len(lumi_g),x,y)
    gr.SetName(f"{name}_lumi")
    # TCanvas.Update() draws the frame, after which one can change it
    gr.SetTitle("")
    gr.SetMarkerStyle( 3 )
    gr.GetYaxis().SetTitle( 'Chamber HV current (Imon)' )
    gr.GetXaxis().SetTitle( 'Instantaneous Luminosity' )
    gr.Draw( 'AP' )
    c1.Draw()

    myfile = rt.TFile( f'../root/chambers.root', 'UPDATE' )
    gr.Write()
    myfile.Close()


    return lumi_g,iMon_g,lumiBlock
    
    
    
```
